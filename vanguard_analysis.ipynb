```python
df_web_exp = pd.merge(df_web, df_experiment, on='client_id', how='inner')
```

```python
df_web_exp.head(120-140)
```

```python
df_counts = df_web_exp.groupby(['Variation', 'visit_id', 'process_step']).size().reset_index(name='step_count')

confusio_mitjana = df_counts.groupby('Variation')['step_count'].mean()
print("Media de repeticiones por paso (más alt = máss confusión):")
print(confusio_mitjana)
```

```python
df_web_exp = df_web_exp.sort_values(['visit_id', 'date_time'])

df_web_exp['prev_step'] = df_web_exp.groupby('visit_id')['process_step'].shift(1)

def is_backtracking(row):
    steps_order = {'start': 0, 'step_1': 1, 'step_2': 2, 'step_3': 3, 'confirm': 4}
    if pd.isna(row['prev_step']): return False
    return steps_order.get(row['process_step'], 0) < steps_order.get(row['prev_step'], 0)

df_web_exp['goes_back'] = df_web_exp.apply(is_backtracking, axis=1)

backtrack_rate = df_web_exp.groupby('Variation')['goes_back'].mean() * 100
print(f"porcentaje de clics para atràs en steps':\n{backtrack_rate}")
```

```python
total_usuarios_grupo = df_web_exp.groupby('Variation')['client_id'].nunique()
confirmacion_por_grupo = df_web_exp[df_web_exp['process_step'] == 'confirm'].groupby('Variation')['client_id'].nunique()
print(confirmacion_por_grupo)
```

```python
confirmacion_rate = (confirmacion_por_grupo / total_usuarios_grupo) * 100
print(confirmacion_rate)
```

```python
df_web_exp['date_time'] = pd.to_datetime(df_web_exp['date_time'])

df_web_exp = df_web_exp.sort_values(['visit_id', 'date_time'])

df_web_exp['time_diff'] = df_web_exp.groupby('visit_id')['date_time'].diff().shift(-1)

df_web_exp['time_diff_seconds'] = df_web_exp['time_diff'].dt.total_seconds()

df_time = df_web_exp[(df_web_exp['time_diff_seconds'] > 0) & (df_web_exp['time_diff_seconds'] < 3600)].copy()

print(f"Tiempo medio por paso")
print(df_time.groupby(['Variation', 'process_step'])['time_diff_seconds'].mean())
```

```python
funnel_data = df_web_exp.groupby(['Variation', 'process_step'])['client_id'].nunique().unstack()

ordre_pas = ['start', 'step_1', 'step_2', 'step_3', 'confirm']
funnel_data = funnel_data[ordre_pas]

print("usuarios únicos por paso")
print(funnel_data)
```

```python

funnel_retention = funnel_data.div(funnel_data['start'], axis=0) * 100

print("Rate de retención")
print(funnel_retention.round(2))
```
